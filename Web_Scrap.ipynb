```python
cities = [{"name":"Hayward","state":"CA","start_page":1,"end_page":9}]
listing_urls = {}
for city_dict in cities:
    homes = []
    base_url = "https://www.realtor.com/realestateandhomes-search/"
    for i in range(city_dict.get("start_page"),city_dict.get("end_page")):
        url = base_url+city_dict.get("name")+'_'+city_dict.get("state")+'/'
        req = urllib.request.Request(url+'pg-'+str(i),headers=headers)
        htmlfile = urlopen(req)
        soup = BeautifulSoup(htmlfile,"html.parser")

        homecards = soup.find_all("li",{"class":"component_property-card"})
        for homecard in homecards:
            a = homecard.find("a")
            if a and a.has_attr('href'):
                url = a["href"]
                if url and city_dict.get("name") in url:
                    homes.append(url)
    listing_urls[city_dict.get("name")] = homes

# print(listing_urls)
```

```python
def findValuePropertyFacts(soup,key):
    lis = soup.find_all("li")
    value_div = ''
    key_span = ''
    for li in lis:
        first_div = li.find("div")
        divs = first_div.find_all("div")
        for div in divs:
            span = div.find("span")
            if span:
                key_span = span
                continue
            value_div = div
        if key_span and value_div and key_span.text==key:
            return value_div.text
    return ""

def findValueNeighborhoodData(soup,key):
    lis = soup.find_all("li")
    for li in lis:
        p = li.find_all("p")
        if p[0].text == key:
            return p[1].text
        elif p[1].text == key:
            return p[0].text

def findBrokerCompany(soup):
    outer_spans = soup.find_all("span")
    company = "-"
    for span in outer_spans:
        if span.has_attr("class") and "desktop-bold" in span.get("class"):
            company = span.text.strip()
    return company
                


base_url = "https://www.realtor.com"
all_dataframes = []
for city,urls in listing_urls.items():
    
    listing_data = []
    school_data = []
    historic_data = []
    
    for i in range(0,20):
        listing_details = {}
        school_details = {}
        historic_details = {}
        
        url = base_url+urls[i]
        
  
        # prints a random value from the list
        list1 = [1]
        time.sleep(random.choice(list1))
        
        req = urllib.request.Request(url,headers=headers)
        htmlfile = urlopen(req)
        soup = BeautifulSoup(htmlfile,"html.parser")
        
        #Url
        listing_details["url"] = urls[i]
        
        #City
        listing_details["city"] = city
        
        #State
        listing_details["state"] = "CA"
    
        #Street Address
        address = "-"
        div = soup.find("div",{"class":"address-value"})
        h1 = ''
        if div:
            h1 = div.find("h1")
        if h1:
            address = h1.text.strip()
            tokens = address.split(",")
            if len(tokens)>0:
                listing_details["street-address"] = tokens[0]
            else:
                listing_details["street-address"] = "-"
        
        #zipcode
        finds = []
        zip = "-"
        if h1:
            zip_regex = r"[0-9]{5}"
            tokens = address.split(",")
            for token in tokens:
                if re.search(zip_regex,token):
                    finds = re.findall(zip_regex,token)
                    if len(finds)>0:
                        zip = finds[0]                
        listing_details["zipcode"] = zip
        
        
        #beds
        beds = "-"
        li = soup.find("li",{"data-testid":"property-meta-beds"})
        if li:
            span = li.find("span")
            if span:
                beds = span.text
        listing_details["beds"] = beds
        
        #baths
        baths = "-"
        li = soup.find("li",{"data-testid":"property-meta-baths"})
        if li:
            span = li.find("span")
            if span:
                baths = span.text
        listing_details["baths"] = baths
        
        #sqft
        sqft = "-"
        li = soup.find("li",{"data-testid":"property-meta-sqft"})
        if li:
            span = li.find("span")        
            if span:
                sqft = span.text
        listing_details["sqft"] = sqft
        
        #lot-size
        lotsize = "-"
        li = soup.find("li",{"data-testid":"property-meta-lot-size"})
        if li:
            span = li.find("span")
            if span:
                lotsize = span.text
        listing_details["lotsize"] = lotsize
        
        #price
        price = "-"
        div1 = soup.find("div",{"data-testid":"list-price"})
        if div1:
            div2 = div1.find("div")        
            if div2:
                price = div2.text
        listing_details["price"] = price
        
        #propertyfacts
        div = soup.find("div",{"data-testid":"listing-indicator"})
        if div:
            ul = div.find("ul")
            if ul:
                #Property Type
                listing_details["property-type"] = findValuePropertyFacts(ul,"Property Type")
                #Time on realtor.com
                listing_details["time-on-realtor"] = findValuePropertyFacts(ul,"Time on realtor.com")
                #HOA Fees
                listing_details["hoa"] = findValuePropertyFacts(ul,"HOA Fees")
                #Price per sqft
                listing_details["price/sqft"] = findValuePropertyFacts(ul,"Price per sqft")
                #Garage
                listing_details["garage"] = findValuePropertyFacts(ul,"Garage")
                #Year Built
                listing_details["year"] = findValuePropertyFacts(ul,"Year Built")
        
        #NeighborhoodData
        div = soup.find("div",{"data-testid":"neighborhood-home-values"})
        if div:
            ul = div.find("ul")
            if ul:
                #Median Listing Home Price
                listing_details["median_listing_home_price"] = findValueNeighborhoodData(ul,"Median Listing Home Price")
                #Median Sold Home Price
                listing_details["median_sold_home_price"] = findValueNeighborhoodData(ul,"Median Sold Home Price")
                #Median Days on Market
                listing_details["median_days_on_market"] = findValueNeighborhoodData(ul,"Median Days on Market")
                #Median Listing Home Price Per Sq Ft
                listing_details["median_listing_home_price_persqft"] = findValueNeighborhoodData(ul,"Median Listing Home Price Per Sq Ft")
        
        
        #schooldistrict
        schooldist = "-"
        li = soup.find("li",{"class":"mls-school-list-row"})
        if li:
            divs = li.find_all("div")
            
            if divs and "District" in divs[0].text:
                schooldist = divs[1].text
            elif divs and "District" in divs[1].text:
                schooldist = divs[0].text
        listing_details["school_district"] = schooldist
        
        #Agent
        company = "-"
        brandInfoDiv = soup.find("div",{"class":"brand-info"})
        if brandInfoDiv:
            #select span which contains presented by
            company = findBrokerCompany(brandInfoDiv)
        listing_details["broker"] = company
        
        
        listing_data.append(listing_details.copy())
        
    all_dataframes.append(pd.DataFrame(listing_data))
```
